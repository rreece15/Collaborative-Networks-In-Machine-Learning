```python
import pandas as pd
import csv
```

```python
_title = 'TI'
_authors = 'AU'
_publication_type = 'PT'
_authors_full = 'AF'
_source_title = 'SO'
_document_type = 'DT'
_keywords = 'DE'
_addresses = 'C1'
_reprint_addresses = 'RP'
_email = 'EM'
_number_references = 'NR'
_times_cites = 'TC'
_publication_date = 'PD'
_publication_year = 'PY'
_volume = 'VL'
_article_number = 'AR'
_doi = 'DI'
_wos_categories = 'WC'
_export_date = 'DA'
_abstract = 'AB'
_cited_references = 'CR'
_book_editors = 'BE'
_end_file = 'EF'
```

```python
def parse_file(file_name):
    file = open("../data/Custom/"+file_name)
    rows = file.readlines()
    file.close()

    df = pd.DataFrame(columns=['Title'])
    counter = 0
    vals = {}
    for line in rows:
        if line[:2] == _end_file:
            break
        elif line == '\n':
            # append existing source and create new source
            vals['Class']=file_name[0]
            df = df.append(vals,ignore_index=True)
            vals.clear()
            # break
        elif line[:2] == _title:
            vals['Title'] = line[2:].strip()
        elif line[:2] == _authors:
            vals['Authors'] = line[2:].strip()
        elif line[:2] == _publication_type:
            vals['Publication Type'] = line[2:].strip()
        elif line[:2] == _authors_full:
            vals['Authors (Full)'] = line[2:].strip()
        elif line[:2] == _source_title:
            vals['Source Title'] = line[2:].strip()
        elif line[:2] == _document_type:
            vals['Document Type'] = line[2:].strip()
        elif line[:2] == _keywords:
            vals['Keywords'] = line[2:].strip()
        elif line[:2] == _addresses:
            vals['Addresses'] = line[2:].strip()
        elif line[:2] == _reprint_addresses:
            vals['Reprint Addresses'] = line[2:].strip()
        elif line[:2] == _email:
            vals['Email'] = line[2:].strip()
        elif line[:2] == _number_references:
            vals['Number References'] = line[2:].strip()
        elif line[:2] == _times_cites:
            vals['Times Cited'] = line[2:].strip()
        elif line[:2] == _publication_date:
            vals['Publication Date'] = line[2:].strip()
        elif line[:2] == _publication_year:
            vals['Publication Year'] = line[2:].strip()
        elif line[:2] == _volume:
            vals['Volume'] = line[2:].strip()
        elif line[:2] == _article_number:
            vals['Article Number'] = line[2:].strip()
        elif line[:2] == _doi:
            vals['DOI'] = line[2:].strip()
        elif line[:2] == _wos_categories:
            vals['WOS Categories'] = line[2:].strip()
        elif line[:2] == _export_date:
            vals['Export Date'] = line[2:].strip()
        elif line[:2] == _abstract:
            vals['Abstract'] = line[2:].strip()
        elif line[:2] == _cited_references:
            vals['Cited References'] = line[2:].strip()
        elif line[:2] == _book_editors:
            vals['Book Editors'] = line[2:].strip()
        # print(line[:2])
    return df
```

```python
neuromorphic = parse_file("Neuromorphic_1.txt")
neuromorphic = neuromorphic.append(parse_file("Neuromorphic_2.txt"),ignore_index=True)
display(neuromorphic)
edge = parse_file("EdgeML_1.txt")
edge = edge.append(parse_file("EdgeML_2.txt"),ignore_index=True)
accelerator = parse_file("MLAccelerator_1.txt")
accelerator = accelerator.append(parse_file("MLAccelerator_2.txt"),ignore_index=True)
spiking = parse_file("SpikingNN_1.txt")
spiking = spiking.append(parse_file("SpikingNN_2.txt"),ignore_index=True)
```

```python
custom_dataset = pd.DataFrame()
custom_dataset = custom_dataset.append(neuromorphic)
custom_dataset = custom_dataset.append(edge)
custom_dataset = custom_dataset.append(accelerator)
custom_dataset = custom_dataset.append(spiking)
display(custom_dataset)
```

```python
custom_dataset.to_csv("../data/Custom/combined.csv")
```
