```python
_title = 'TI'
_authors = 'AU'
_publication_type = 'PT'
_authors_full = 'AF'
_source_title = 'SO'
_document_type = 'DT'
_keywords = 'DE'
_addresses = 'C1'
_reprint_addresses = 'RP'
_email = 'EM'
_number_references = 'NR'
_times_cited = 'TC'
_publication_date = 'PD'
_publication_year = 'PY'
_volume = 'VL'
_article_number = 'AR'
_doi = 'DI'
_wos_categories = 'WC'
_export_date = 'DA'
_abstract = 'AB'
_cited_references = 'CR'
_book_editors = 'BE'
_end_file = 'EF'
_end_row = 'ER'

_title_column = 'Title'
_authors_column = 'Authors'
_publication_type_column = 'Publication Type'
_authors_full_column = 'Authors (Full)'
_source_title_column = 'Source Title'
_document_type_column = 'Document Type'
_keywords_column = 'Keywords'
_addresses_column = 'Addresses'
_reprint_addresses_column = 'Recipient Addresses'
_email_column = 'Email'
_number_references_column = 'Number References'
_times_cited_column = 'Times Cited'
_publication_date_column = 'Publication Date'
_publication_year_column = 'Publication Year'
_volume_column = 'Volumn'
_article_number_column = 'Article Number'
_doi_column = 'DOI'
_wos_categories_column = 'WOS Categories'
_export_date_column = 'Export Date'
_abstract_column = 'Abstract Column'
_cited_references_column = 'Cited References'
_book_editors_column = 'Book Editors'

prefixes = {
    _title: _title_column,
    _authors: _authors_column,
    _publication_type: _publication_type_column,
    _authors_full: _authors_full_column,
    _source_title: _source_title_column,
    _document_type: _document_type_column,
    _keywords: _keywords_column,
    _addresses: _addresses_column,
    _reprint_addresses: _reprint_addresses_column,
    _email: _email_column,
    _number_references: _number_references_column,
    _times_cited: _times_cited_column,
    _publication_date: _publication_date_column,
    _publication_year: _publication_year_column,
    _volume: _volume_column,
    _article_number: _article_number_column,
    _doi: _doi_column,
    _wos_categories: _wos_categories_column,
    _export_date: _export_date_column,
    _abstract: _abstract_column,
    _cited_references: _cited_references_column,
    _book_editors: _book_editors_column,
    'BP': 'idk',
    'EP': 'idk',
    'VL': 'Volume',
    'Z9': 'Total Cited (all db)',
    'U1': '180 Usage',
    'U2': 'Usage Since 2013',
    'IS': 'Issue',
    'D2': 'DOI2',
}
```

```python
def parse_file(file_name):
    file = open("../data/Custom/"+file_name)
    rows = file.readlines()
    file.close()
    vals = {}
    df = pd.DataFrame(columns=['Title'])
    last_state = _end_row
    counter = 0
    state_array = []
    for line in rows:
        prefix = line[:2]
        if prefix == _end_row: # add the row to the dataframe
            vals['Class']=file_name[0]
            # print(vals)
            df = df.append(vals,ignore_index=True)
            state_array.clear()
            vals.clear()
            # if counter > 1:
            #     break
            # counter = counter + 1
            continue
        if prefix in prefixes.keys():
            if not last_state == _end_row:
                if last_state == _title:
                    state_array = [''.join(state_array)]
                if len(state_array) == 1:
                    vals[prefixes[last_state]] = state_array[0]
                else:
                    vals[prefixes[last_state]] = state_array.copy()
                state_array.clear()
            last_state = prefix
            if prefix == _cited_references:
                try:
                    line = prefix + line.split(' DOI ')[1].strip()
                except(IndexError):
                    line = line
            state_array.append(line[2:].strip())
        else:
            if last_state == _cited_references:
                try:
                    line = line.split(' DOI ')[1].strip()
                except(IndexError):
                    line = line
            state_array.append(line.strip())
    return df
```

```python
neuromorphic = parse_file("Neuromorphic_1.txt")
neuromorphic = neuromorphic.append(parse_file("Neuromorphic_2.txt"),ignore_index=True)
display(neuromorphic)
edge = parse_file("EdgeML_1.txt")
edge = edge.append(parse_file("EdgeML_2.txt"),ignore_index=True)
accelerator = parse_file("MLAccelerator_1.txt")
accelerator = accelerator.append(parse_file("MLAccelerator_2.txt"),ignore_index=True)
spiking = parse_file("SpikingNN_1.txt")
spiking = spiking.append(parse_file("SpikingNN_2.txt"),ignore_index=True)
```

```python
custom_dataset = pd.DataFrame()
custom_dataset = custom_dataset.append(neuromorphic)
custom_dataset = custom_dataset.append(edge)
custom_dataset = custom_dataset.append(accelerator)
custom_dataset = custom_dataset.append(spiking)
custom_dataset = custom_dataset.reset_index()
custom_dataset  = custom_dataset.drop('index', axis=1)
display(custom_dataset)
```
