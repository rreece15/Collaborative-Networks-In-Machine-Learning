## Look at papers per year per category

```python
import pandas as pd

df = pd.read_csv('wos.complete.connected.csv')

# extract a list of years
years = df['year'].tolist()
categories = df['category'].tolist()
```

```python
neuro = "Neuromorphic"
spiking = "SpikingNN"
mlacc = "MLAccelerator"
edge = "EdgeML"
```

```python
neurolist = []
spikinglist = []
mlacclist = []
edgelist = []
```

```python
for i in range(len(categories)):
    if categories[i] == neuro:
        neurolist.append(years[i])
    elif categories[i] == spiking:
        spikinglist.append(years[i])
    elif categories[i] == mlacc:
        mlacclist.append(years[i])
    elif categories[i] == edge:
        edgelist.append(years[i])
```

```python
print(edgelist)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import *

# binwidth = 5000
plt.hist(neurolist, bins=np.arange(1970, 2022, 1))
plt.ylim(ymin=0, ymax = 1400)
plt.title("Neuromorphic Computing Number of papers vs year")
plt.xlabel("Year")
plt.ylabel("Occurrences")
```

```python
# binwidth = 5000
plt.hist(spikinglist, bins=np.arange(1970, 2022, 1))
plt.ylim(ymin=0, ymax = 1400)
plt.title("Spiking NN Number of papers vs year")
plt.xlabel("Year")
plt.ylabel("Occurrences")
```

```python
# binwidth = 5000
plt.hist(mlacclist, bins=np.arange(1970, 2022, 1))
plt.ylim(ymin=0, ymax = 1400)
plt.title("ML Accelerators Number of papers vs year")
plt.xlabel("Year")
plt.ylabel("Occurrences")
```

```python
# binwidth = 5000
plt.hist(edgelist, bins=np.arange(1970, 2022, 1))
plt.ylim(ymin=0, ymax = 1400)
plt.title("ML on Edge Number of papers vs year")
plt.xlabel("Year")
plt.ylabel("Occurrences")
```
