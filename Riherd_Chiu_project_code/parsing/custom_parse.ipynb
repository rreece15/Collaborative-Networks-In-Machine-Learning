## parsing Function and Main

```python
def parse_file(file_name):
    file = open("../data/updated_wos/"+file_name)
    rows = file.readlines()
    file.close()
    vals = {}
    df = pd.DataFrame(columns=['Title'])
    last_state = _end_row
    counter = 0
    state_array = []
    for line in rows:
        prefix = line[:2]
        if prefix == _end_row: # add the row to the dataframe
            vals['Class']=file_name[0]
            # print(vals)
            df = df.append(vals,ignore_index=True)
            state_array.clear()
            vals.clear()
            # if counter > 1:
            #     break
            # counter = counter + 1
            continue
        if prefix in prefixes.keys():
            if not last_state == _end_row:
                if last_state == _title:
                    state_array = [''.join(state_array)]
                if len(state_array) == 1:
                    vals[prefixes[last_state]] = state_array[0]
                else:
                    vals[prefixes[last_state]] = state_array.copy()
                state_array.clear()
            last_state = prefix
            if prefix == _cited_references:
                try:
                    line = prefix + line.split(' DOI ')[1].strip()
                except(IndexError):
                    line = line
            state_array.append(line[2:].strip())
        else:
            if last_state == _cited_references:
                try:
                    line = line.split(' DOI ')[1].strip()
                except(IndexError):
                    line = line
            state_array.append(line.strip())
    return df
```
